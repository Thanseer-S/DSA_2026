```python
import pandas as pd

df = pd.read_csv('/content/Sales_add.csv')

df.head()
```

```python
from scipy.stats import ttest_rel

before = df['Sales_before_digital_add(in $)']
after = df['Sales_After_digital_add(in $)']

t_stat, p_value = ttest_rel(before, after)

print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Significant increase in sales after digital marketing.")
else:
    print("No significant increase in sales.")
```

```python
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df['Region'], df['Manager'])

print("Contingency Table:\n", contingency_table)

chi2, p, dof, expected = chi2_contingency(contingency_table)

print("\nChi-square value:", chi2)
print("P-value:", p)

if p < 0.05:
    print("Conclusion: Region and Manager are dependent.")
else:
    print("Conclusion: Region and Manager are independent.")
```
